```python
from mysklearn.mypytable import MyPyTable
import random

table = MyPyTable().load_from_file("input_data/tracks.csv")

def categorize_popularity(p):
    p = int(p)
    if p <= 33:
        return "Low"
    elif p <= 66:
        return "Medium"
    else:
        return "High"

low_idx = []
med_idx = []
high_idx = []

pop_idx = table.column_names.index("popularity")

for i, row in enumerate(table.data):
    pop_labels = categorize_popularity(row[pop_idx])
    pop_labels_idx = i

    if pop_labels == "Low":
        low_idx.append(i)
    elif pop_labels == "Medium":
        med_idx.append(i)
    else:
        high_idx.append(i)

min_count = len(high_idx)

low_sample = random.sample(low_idx, min_count)
med_sample = random.sample(med_idx, min_count)
high_sample = random.sample(high_idx, min_count)

final_idx = low_sample + med_sample + high_sample

print(f"Numbers of rows to keep: {len(final_idx)}")

all_indexes = set(range(len(table.data)))
keep_indexes = set(final_idx)

rows_to_drop = list(all_indexes - keep_indexes)

print(f"Numbers of rows to drop: {len(rows_to_drop)}")

print(f"Before clean: {table.get_shape()}")

table.drop_rows(rows_to_drop)

print(f"After clean: {table.get_shape()}")

table.save_to_file("cleaned_tracks.csv")
```
